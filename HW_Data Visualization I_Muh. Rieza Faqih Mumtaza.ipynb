# 1.&nbsp; Count of one categorical feature

```python
sns.set(color_codes=True)
sns.histplot(df['Fence'], 
             color = 'red')
plt.title('Penggunaan Pagar dalam Perumahan')
plt.ylabel('Frequency')
plt.xlabel('Jenis Pagar');
```

Dalam data tersetbut dapat dilihat bahwa penggunaan pagar MnPrv sangat mendominasi pada setiap rumah yang berjumlah hampir 160, sedangkan pagar GdWo dan GdPrv memiliki jumlah yang hampir mirip pada range 50-60. Penggunaan pagar yang terendah yaitu MnWw berada di angka 20 kurang. Hal ini menunjukkan terdapat perbedaan yang cukup signifikan antara penggunaan pagar MnPrv dengan pagar yang lainnya dan dapat dikatakan bahwa pemilihan pagar MnPrv adalah pagar yang cukup favorit digunakan pada rumah-rumah.

# 2.&nbsp; Distribution of one continuous feature

```python
sns.displot(df['MoSold'], 
             color = 'brown',
             kde = True) 
median_value = np.median(df['MoSold'])
plt.axvline(median_value, color='red', linewidth=2, label='Median')
plt.title('Penjualan Tiap Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah')
```

Dari data tersebut dapat dilihat bahwa data terdistribusi tidak normal dengan skew ke kanan dan memiliki median diangka 250 pada bulan ke 6. Penjualan terendah ada pada bulan kedua dan tertinggi ada pada bulan ke 6, dari data tersebut juga dapat dikatakan bahwa penjualan disetiap bulan mengalami ketidakstabilan disebabkan kenaikan dan penurunannya.

# 3.&nbsp; Categorical-continuous

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='SalePrice', y='LotShape', data=df)
plt.title('Hubungan Antara Sale Price dan Lot Shape')
plt.xlabel('SalePrice')
plt.ylabel('LotShape')
plt.show()
```

Hubungan Antara Harga Jual dan Bentuk Box Plot:

Kecenderungan: Pada umumnya, tidak ada kecenderungan yang jelas antara bentuk lot dan harga jual. Hal ini terlihat dari median (garis di tengah kotak) yang hampir sama untuk semua bentuk lot.

Variasi: Variasi harga jual untuk setiap bentuk lot cukup besar. Hal ini terlihat dari kotak yang lebar dan kumis yang panjang.

Outliers: Ada beberapa outliers (titik di luar kumis) untuk setiap bentuk lot. Outliers ini menunjukkan bahwa ada beberapa lot dengan harga jual yang jauh lebih tinggi atau lebih rendah daripada harga jual rata-rata untuk bentuk lot tersebut.

Analisis Lebih Detail:

Berikut adalah analisis lebih detail untuk setiap bentuk lot:

1. Regular:
Variasi harga jual: cukup besar (kotak lebar)
Outliers: ada beberapa outliers di atas dan di bawah median
2. IR1:
Variasi harga jual: cukup besar (kotak lebar)
Outliers: ada beberapa outliers di atas median
3. IR2:
Variasi harga jual: paling kecil (kotak paling sempit)
Outliers: ada beberapa outliers
4. IR3:
Variasi harga jual: cukup besar (kotak lebar)
Outliers: tidak ada beberapa outliers di bawah median

Berdasarkan analisis boxplot, dapat disimpulkan bahwa:
1.Tidak ada kecenderungan yang jelas antara bentuk lot dan harga jual.
2.Variasi harga jual untuk setiap bentuk lot cukup besar.
3.Ada beberapa outliers untuk setiap bentuk lot.

# 4.&nbsp; Continuous-continuous

```python
sns.regplot(
    data=df,
    x='LotFrontage',
    y='SalePrice',
    scatter_kws = {'alpha' : 0.5},
    line_kws = {'color' : 'red'}
);
```

Hubungan Antara Lot Frontage dan Sale Price:

Kecenderungan: Ada hubungan positif antara lot frontage dan sale price. Hal ini terlihat dari garis merah yang miring ke atas.
Kekuatan: Hubungan antara lot frontage dan sale price cukup kuat
Arah: Semakin luas lot frontage, semakin tinggi sale price.

Garis merah: Garis merah represents the least squares regression line. Garis ini menunjukkan nilai sale price yang diprediksi untuk setiap nilai lot frontage.
Titik-titik: Titik-titik pada plot menunjukkan data aktual untuk sale price dan lot frontage.

Berdasarkan analisis regresi plot, dapat disimpulkan bahwa:

Ada hubungan positif yang kuat antara lot frontage dan sale price.
Semakin luas lot frontage, semakin tinggi sale price.
Garis regresi menunjukkan nilai sale price yang diprediksi untuk setiap nilai lot frontage.
